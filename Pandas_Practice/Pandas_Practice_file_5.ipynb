```python
df=pd.DataFrame({'a':[1,2,3,4,5,6,7,8],'b':[10,np.nan,np.nan,20,np.nan,np.nan,30,30]})
```

```python
df['a_cumsum']=df.a.cumsum()
df
```

```python
obj=df['a'].rolling(window=2)
```

```python
df['a_roll_sum']=obj.sum()
```

```python
df
```

```python
obj=df['b'].rolling(window=2,min_periods=2)
```

```python
df['b_roll_sum_count']=obj.count()
df
```

```python
obj1=df['a'].expanding(min_periods=1)
```

```python
df['a_exp_sum']=obj1.sum()
df
```

```python
df['a_exp_mean']=obj1.mean()
```
